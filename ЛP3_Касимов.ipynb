Реализуем чтение всего файла в строку с использованием метоа `open()` и `read()`:

```python
f=open('data_v1.txt')
s=f.read().strip()
print(len(s))
```

## Решение задачи

```python
target=''
maxx=0
maxs=''
for c in s:
  if c.isdigit() and int(c)%2!=0:
    target+=str(c)
  else:
    if len(target)>maxx:
      maxx=len(target)
      maxs=target
      target=''

print('Максимальная длина подпоследовательности нечетных чисел = %d' % maxx)
print('Подпоследовательность:', maxs)
```
